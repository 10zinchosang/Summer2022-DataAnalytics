## Data

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

**Import matplotlib.pyplot as plt and set %matplotlib inline if you are using the jupyter notebook. What command do you use if you aren't using the jupyter notebook?**

```python
import matplotlib.pyplot as plt
```

## Exercise 1

* **Create a figure object called fig using plt.figure() **
* **Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
* **Plot (x,y) on that axes and set the labels and titles to match the plot below: **

```python
# Create a figure
fig = plt.figure()

# Create a set of axes that we can plot on
ax = fig.add_axes([0, 0, 1, 1])

# Plot and set labels
ax.plot(x, y, 'b')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_title('title');

# Set the "window" of the graphs and faces tickmarks inwards
ax.set_xlim([0, 100])
ax.set_ylim([0, 200])
ax.tick_params(direction = 'in')

# Set tickmarks
ax.set_xticks([0,20,40,60,80,100])
ax.set_yticks([0,50,100,150,200]);
```

## Exercise 4

**Use plt.subplots(nrows=1, ncols=2) to create the plot below.**

```python
# Create two plots horizontally
fig,axes = plt.subplots(nrows=1,ncols=2)
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
# Plot with different line styles
axes[0].plot(x,y, color='b', linewidth=3, ls='--')
axes[1].plot(x,z, color='r', linewidth=3)

# Set "window" and ticks if needed
axes[0].set_xlim([0, 100])
axes[0].set_ylim([0, 200])
axes[0].set_yticks([0,50,100,150,200])

axes[1].set_xlim([0, 100])
axes[1].set_ylim([0, 10000])

# Call fig to see
fig
```

*See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.*

```python
# Create fig and change figsize
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(12,2))

# Plot with different linestyles
axes[0].plot(x,y, color='b', linewidth=4)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')

axes[1].plot(x,z, color='r', linewidth=3, ls='--')
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')

# Set "window" and ticks if needed
axes[0].set_xlim([0, 100])
axes[0].set_ylim([0, 200])
axes[0].set_yticks([0,50,100,150,200])

axes[1].set_xlim([0, 100])
axes[1].set_ylim([0, 10000]);
```
